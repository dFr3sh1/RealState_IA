### Types de Formes de Lots:
    
    On a 5 principaux:
    
    1- Rectangle: Ils sont courantes dans des cities comme Edmonton
    "Les pros"
        - Plus facile à mesurer, par conséquence, plus faciles à délimiter les lignes de la propriété.
        - Ils fournissent une meilleure exploitation de la surface.
        - Plus facile et moins couteuses à protéger
        - Plus faciel pour distribuer les fronts des arrières et les jardins
        - Ils conservent plus la valeurs par rapports aux lots avvec des formes irregulières.
    
    2- Carré À différence des premiers les jardins son plus proches les bénéfices restent presque les mêmes.

    3- Pie-Shaped Lots: Très communs surtout qu'il y a bcp de communautés qui construisent les quartiers en cul-de-sac pour gagenr en intimité pour éviter le traffic de voitures non voisins.
    Généralment sont classifiés dans la catégorie de formes triangulairs.
    "Les Pros"
    - Ils proposent de vastes jardins
    - Les limites sont faciles à maintenir.
    - Rélativement facile à lever des barrières
    - Les surfaces offrent la possibilité de faire des bbq et des piscines

    "Les cons"
    - Une façade plus étroite et moins de places de parking
    - Surfaces plus difficile à exploiter.

    4- Reverse Pie-shaped lot : À l'inverse des précédents les jardins sont plus étroits
    "Les pros"
    - Façades plus vastes
    - Plus d'espaces pour parking
    - Plus attractives de l'extérieur
    - Limites faciles à entreténir

    "Les cons"
    - Jardins plus petits
    - Plus difficiles à conserver la valeur car les acheteurs veulent des grarnds jardins
    - Moins d'intimité car la façade plus large.
    - Plus difficile à exploiter car la surface esst trop irregulière.

   5- Mots de formes irregulières:
    Ils n'appartient à aucun des styles mentionnés précédemment.
    "Les pros"
    - Opportunité de construire maisons uniques
    - Mois chers par mètre carré
    - Ses surfaces irregulères peuvent finir en piscines, zones de jeux et jardins

    "Les cons"
    - Trop difficile à exploiter
    - Plus difficiles à protéger
    - Plus difficile à délimiter
    - Difficiel à conserer leur valeur car les acheteurs les fuient.

### Les types de positionnement

Le positionnement des lots jeu un rôle fondamental car ils affectent le valeur de la propriété.

Les acheteurs veulent acheter des lots qui garantissent un environnement apaisé, espacieux, intimes et avec de places de parking

Les plus connus sont :

**Cul-de-sac**

Ils sont positionnés à la fin des rues ou des rues cul-de-sac, d'où son nom.
"Les pros"
- Plus appréciés car moins de traffic des non riverains
- Jardins bien larges
- Environnements plus calmes
- Ils gagnent en valeur grâce à leur attributs de privacité t intimité.

"Les cons"
- Façades plus etroites
- Moins d'attractivité depuis l'extérieur.

**Lots d'angle**

Bien connus par leur positionnement entre l'intersection des deux rues.
"les pros"
- Facile accès
- Plus d'attracitivé depuis l'extérieur

"Les cons"
- Moins de privacité à cause du traffic de voitures.
- Plus bruyantes
- Se déprécient plus rapidement à cause du bruyard.

**Key Lots**

Lots qui sont bordés pour plus de rois lots autour sauf le front, ce qui donne l'impression d'être encloisonés.
"Les cons"
- Moins de privacité
- Plus ou moins bruyard
- Dépréciation de la valeur

**Lots d'intérieur**

Ils sont les plus populaires. Ils entourés par d'autres lots par les côtés lateraux. 
"Les pros"
- Ils sont beaucoup appréciés.
- Ils gagnent en valeur grâce à son poupularité
- Ils sont des surfaces appropriés, généralement carrrés ou rectangulaires.

les "cons"
- Moins de privacité car ils sont de voisins sur les deux côtés.


**Les Intersection en T**

Ils sont situés en face d'un intersection des deux rues.
"Les cons"
- Ces lots suffrent bcp du bruyard et du  manque d'intimité.
- Haut rique d'accidents à cause de voitures qui vient par le front.
- Dépréciation de la valeur

**Les Lots drapeaux** (Je n'ai pas trouvé l'équivalent en frances, peut-être "cours")

Les types de lots qui pareilsent cachés. Ils peuvent être accédes par des petits accès car ils sont situés derrier d'autres lots.

"Les cons"
- Pas d'intimité car ils ont entoures par tous les côtés.
- Pas d'attractivité despuis l'extérieur
- Moins appréciation de la valeur.
